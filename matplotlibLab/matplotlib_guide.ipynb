### Step 3: draw bar graph

```python
import matplotlib.pyplot as plt
import numpy as np

# fix x, y
x = np.array([2018,2019,2020,2021,2022])
#y = np.array([1050, 2452.20, 3456.30, 4653.40, 4052.60])
y = np.random.randint(100, 5001, 5)

plt.bar(x,y, color=['green','red','blue','yellow','pink'], edgecolor='black')
plt.show()
```
